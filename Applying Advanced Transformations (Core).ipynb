### Loading Data 

```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
df = pd.read_csv('Data/superhero_info - superhero_info.csv')
df.head()
```

```python
df.info()
```

```python
df2 = pd.read_csv('Data/superhero_powers - superhero_powers.csv')
df2.head()
```

### Separate Hero and Publisher column

```python
df['Hero|Publisher'].head()
```

```python
#Split hero/publisher
df[['Hero', 'Publisher']] = df['Hero|Publisher'].str.split('|', expand = True)
#Verify the split
df.head(2)
```

```python
#Drop Hero|Publisher
df = df.drop(columns = 'Hero|Publisher')
#Verify the drop
df.head()
```

### Separate Height and Weight in Measurements column

```python
##Split measurements into height and weight
#Use str.replace to replace single quotes
df['Measurements'] = df['Measurements'].str.replace("'",'"')
#Apply json.loads to whole column
df['Measurements'] = df['Measurements'].apply(json.loads)
#Verify changes
df['Measurements'].head(2)
```

```python
#Check a single value after transformation
test_measurement = df.loc[0, 'Measurements']
print(type(test_measurement))
test_measurement
```

```python
#Create series of height and weight values
height_weight = df['Measurements'].apply(pd.Series)
height_weight
```

```python
#Split columns to obtain only numeric values
height_weight[['Height (cm)', 'cm']] = height_weight['Height'].str.split(' ', expand = True)
height_weight[['Weight (kg)', 'kg']] = height_weight['Weight'].str.split(' ', expand = True)

#Verify changes
height_weight
```

```python
#Drop height, weight, cm, and kg columns
height_weight = height_weight.drop(columns = ['Height', 'Weight', 'cm', 'kg'])

#Verify the changes
height_weight.head()
```

### Concatenate DataFrames 

```python
#Concatenate frames
df = pd.concat((df, height_weight), axis = 1)
df.head(2)
```

```python
#Drop measurements column
df = df.drop(columns = 'Measurements')
df.info()
```

```python
#Change height and weight columns to numeric types
df[['Height (cm)', 'Weight (kg)']] = df[['Height (cm)', 'Weight (kg)']].astype('float')
df.info()
```

```python
#Check values of Powers column
test_power = df2.loc[0, 'Powers']
test_power
```

```python
#Create new column where powers are split on the comma
df2['Powers_Split'] = df2['Powers'].str.split(',')
df2['Powers_Split'].head(2)
```

```python
#Explode columns
exploded = df2.explode('Powers_Split')
exploded.head()
```

```python
#Saving the unique values from the exploded column
cols_to_make = exploded['Powers_Split'].dropna().unique()
cols_to_make
```

```python
#Loop through columns to add True/False labels
for col in cols_to_make:
    df2[col] = df2['Powers'].str.contains(col)
df2.head()
```

```python
#Drop powers and powers_split columns
df2 = df2.drop(columns = ['Powers', 'Powers_Split'])
#Verify the change
df2.head()
```

```python
#Merge dataframes into 1
merged = pd.merge(df, df2, left_on = 'Hero', 
                  right_on = 'hero_names', how = 'inner')

merged.head()
```

Compare the average weight of super powers who have Super Speed to those who do not.

```python
#Create visualization of super hero weights with super speed power
ax = sns.barplot(data = merged, x = 'Super Speed', y = 'Weight (kg)')
ax.set_title('Comparison of Average Weight of Super Speed Heroes');
```

What is the average height of heroes for each publisher?

```python
#Compare heights of heroes for each publisher
avg_height = merged.groupby('Publisher')['Height (cm)'].mean()
avg_height
```
